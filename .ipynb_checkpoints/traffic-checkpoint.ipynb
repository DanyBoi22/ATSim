```python
from dataclasses import dataclass, field
from enum import Enum, auto
```

```python
class FrameType(Enum):
    DATA = auto()
    CONTROL = auto()

@dataclass
class Frame:
    frame_id: int
    stream_id: str
    size: int               # bytes
    creation_time: float
    frame_type: FrameType = FrameType.DATA
    metadata: dict = field(default_factory=dict)

@dataclass
class Stream:
    stream_id: str
    source: str
    destination: str
```
