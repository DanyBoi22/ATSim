```python
from ipynb.fs.full.traffic import Frame
import pandas as pd
```

```python
class Logger:
    def __init__(self):
        self.records = []

    def log(self, time, node, event, frame: Frame, port):
        self.records.append({
            "time": time,
            "node": node,
            "event": event,
            "frame_id": frame.frame_id,
            "stream_id": frame.stream_id,
            "port": port
        })

    def to_dataframe(self):
        return pd.DataFrame(self.records)
```
